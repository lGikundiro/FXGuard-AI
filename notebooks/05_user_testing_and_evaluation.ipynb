# 05 — User Testing and Evaluation

This notebook helps you organize usability feedback from 5–10 participants.

Participants are not used to train the ML model. They only test the prototype and give feedback on clarity, usefulness, relevance, and usability.

## Suggested participant tasks

Ask each participant to:

1. Open the FXGuard prototype.
2. Select USD/RWF.
3. Enter a sample supplier invoice amount, for example 10,000 USD.
4. Choose 7 days or 14 days.
5. Run risk analysis.
6. Read the risk result.
7. Review the suggested action.
8. Give feedback.

Participants should use hypothetical or estimated values, not confidential business records.

```python
from pathlib import Path
import pandas as pd

ROOT = Path.cwd()
FEEDBACK_DIR = ROOT / 'reports' / 'feedback'
FEEDBACK_DIR.mkdir(parents=True, exist_ok=True)
feedback_file = FEEDBACK_DIR / 'prototype_feedback.csv'
feedback_file
```

## Feedback form fields

Use these fields in Google Forms or in the in-app feedback endpoint:

- participant_id
- role/background
- clarity_rating: 1–5
- usefulness_rating: 1–5
- ease_of_use_rating: 1–5
- recommendation_relevance_rating: 1–5
- would_use_before_supplier_payment: Yes/No/Maybe
- confusing_parts
- suggested_improvements

```python
# Create a blank feedback template if one does not exist.
columns = [
    'participant_id',
    'role_background',
    'clarity_rating',
    'usefulness_rating',
    'ease_of_use_rating',
    'recommendation_relevance_rating',
    'would_use_before_supplier_payment',
    'confusing_parts',
    'suggested_improvements'
]

if not feedback_file.exists():
    pd.DataFrame(columns=columns).to_csv(feedback_file, index=False)

pd.read_csv(feedback_file).head()
```

## Analyze feedback after testing

After collecting responses, run the cells below to summarize average scores and comments.

```python
feedback = pd.read_csv(feedback_file)
feedback.head()
```

```python
if len(feedback) == 0:
    print('No feedback collected yet.')
else:
    rating_cols = [
        'clarity_rating', 'usefulness_rating', 'ease_of_use_rating', 'recommendation_relevance_rating'
    ]
    display(feedback[rating_cols].describe().round(2))
    print('
Would use before supplier payment:')
    print(feedback['would_use_before_supplier_payment'].value_counts(dropna=False))
```

```python
if len(feedback) > 0:
    print('Confusing parts:')
    for item in feedback['confusing_parts'].dropna().tolist():
        print('-', item)
    print('
Suggested improvements:')
    for item in feedback['suggested_improvements'].dropna().tolist():
        print('-', item)
```
